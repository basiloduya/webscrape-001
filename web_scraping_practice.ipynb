```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Africa_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table' , class_ = 'wikitable sortable')
```

```python
soup.find_all('th')
```

```python
africa_titles = soup.find_all('th')
```

```python
africa_titles
```

```python
africa_continent_titles = [title.text.strip() for title in africa_titles]

print(africa_continent_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = africa_continent_titles)
df
```

```python
table = soup.find('table' , class_ = 'wikitable sortable')
```

```python
rows = table.find_all('tr')
```

```python
data = []
for row in rows[1:]:
        cells = row.find_all('td')
        row_data = [cell.text.strip() for cell in cells]
        data.append(row_data)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data, columns=africa_continent_titles)
```

```python
df.to_csv('africa_companies.csv', index=False)
```
